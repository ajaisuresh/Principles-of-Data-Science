```python
import pandas as pd
tp = pd.read_excel("ticketPrices.xlsx")
passengers = pd.read_csv("passengerData.csv")

print(tp)
print(passengers)
```

```python
tpmerge = pd.merge(tp, passengers, how = "outer", left_on = "TicketType", right_on = "p_TicketType")
```

```python
tpmerge
```

```python
tpmerge.info()
```

```python
missing_ages = tpmerge[tpmerge['p_Age'].isnull()]

mean_ages = tpmerge.groupby(['p_Sex','p_Pclass'])['p_Age'].mean()

def remove_na_ages(row):
    
    if pd.isnull(row['p_Age']):
        return mean_ages[row['p_Sex'],row['p_Pclass']]
    else:
        return row['p_Age']

tpmerge['p_Age'] = tpmerge.apply(remove_na_ages, axis=1)
```
